# Q4 What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical). Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.
# Imbalance Dataset:
- An imbalanced dataset refers to a situation in machine learning where the number of observations across different classes is unevenly distributed, meaning one class (the majority class) has far more examples than the other (the minority class)
# Two Technique To Handle Imbalanced Dataset:
# 1. Synthetic Minority Over Sampling Technique(SMOTE):
- SMOTE artificially increases the number of minority class samples by creating synthetic examples, not duplicates. It interpolates new samples along the line segments joining minority class neighbors.
- Practical Examples:
- from imblearn.over_sampling import SMOTE
- from collections import Counter
# Assuming imputed_df contains your features and df["A"] contains your labels
- X, y = imputed_df, df["A"]  # Input features and labels
- print('Original class distribution:', Counter(y))

- smote = SMOTE(random_state=42)
- X_resampled, y_resampled = smote.fit_resample(X, y)
- print('After SMOTE:', Counter(y_resampled))
#2. Random Under/ Oversampling:
- Random Oversampling: Duplicates samples from the minority class to balance the dataset. Simple but may cause overfitting since identical examples are repeated.
- Random Undersampling: Removes random samples from the majority class so classes are balanced. Reduces training time but can lose valuable information.
- Practical Examples:
- from imblearn.over_sampling import RandomOverSampler
- from imblearn.under_sampling import RandomUnderSampler

# Random Oversampling
- ros = RandomOverSampler(random_state=42)
- X_over, y_over = ros.fit_resample(X, y)

# Random Undersampling
- rus = RandomUnderSampler(random_state=42)
- X_under, y_under = rus.fit_resample(X, y)


#Q5 Why is feature scaling important in ML? Compare Min-Max scaling and Standardization. Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient descent.
# Feature Scaling:
- Feature scaling is a data preprocessing technique that transforms numerical features so that they all exist on a comparable scale. It ensures that each feature contributes equally to the learning process, preventing features with large numeric ranges from dominating model behavior.
- Feature scaling is critical for algorithms that rely on distance calculations or gradient-based optimization. Without scaling, features like income (in lakhs) may overpower age (in years) simply due to their magnitude, leading to biased and inefficient models.

# Importance Of Feature Scaling:
- Improves Model Accuracy: Scaling aligns all features to a common range, helping models like logistic regression or neural networks learn optimal weights effectively
- Speeds up Convergence: In gradient descent and similar optimization algorithms, scaling ensures more uniform updates, reducing the risk of oscillations and speeding convergence.
- Equal Contribution of Features: It prevents features with large magnitudes from dominating algorithms that use distance measures, such as K-Nearest Neighbors (KNN) and Support Vector Machines (SVM)
- Enhances Model Interpretability: Scaled features make comparing model coefficients (e.g., in linear regression) or feature importance scores easier and more intuitive
# Impact ON Distance Based Algorithms:
- Distance-Based Algorithms (KNN, SVM, K-Means): Without scaling, features with large ranges dominate Euclidean distance calculations, skewing results and clustering structures.
- Gradient-Based Algorithms (Linear/Logistic Regression, Neural Networks): Scaling improves convergence by preventing drastically uneven parameter updates
- Min-Max Scaling rescales features to , ideal for bounded range models
- Standardization ensures mean-centered features, helping models that assume normalized input.
- Practical Examples:
- from sklearn.preprocessing import MinMaxScaler, StandardScaler
- import pandas as pd

# Sample data
- df = pd.DataFrame({'Age': [18, 25, 40, 60], 'Income': [20000, 50000, 80000, 120000]})

# Min-Max Scaling
- minmax = MinMaxScaler()

# Standardization
- standard = StandardScaler()
- df_standard = standard.fit_transform(df)

#Q6 Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other? Hint: Consider categorical variables with ordinal vs. nominal relationships.
- Label Encoding and One Hot Encoding are the two popular methods used for converting categorical data into numerical form for machine learning but they serve different purpose based on the nature of the categorical variable - whether it is nominal or ordinal
# Label Encoding:
-  Label Encoding assigns a unique integer (or label) to each category. For example, a feature "Color" with categories ["Red", "Blue", "Green"] would be encoded as [0, 1, 2] respectively.
- The advantage is it doesn't increase dimensionality; it keeps data compact. Works well with tree-based models like Decision Trees and Random Forests that can handle ordinal data.
- The disadvantage is it  can mislead models that assume numeric values imply magnitude or order when none exists.
- The nature of data is ordinal
- The output is single integer column
- The dimensionality impact has no increase
- The model suitability has tree based model
- The interpretation has numeric value imply order
- The risk may impose false ordering
# One Hot Encoding :
- One-Hot Encoding creates a binary column for each category of the feature. Each category is represented by a 1 in its respective column and 0 in all others.
- Create a new binary columns for each category with a 1 indicating presence and 0 as absence
- Advantage is prevents the model from inferring a false ordinal relationship good for linear models, neural network and distance based algorithms
- Disadvantage is it increases dimensionality and can lead to sparse data and higher computational demands for variables with many category
- Nature of data is nominal ( that is no order in the data )
- Output is multiple binary columns
- Dimensionality increases one column per category
- Model suitability has linear models, neural networks , KNN
- Interpretation is that each columns represent distinct category
- There is a risk that can cause high dimensional sparse data










```python
#Q7 Google Play Store Dataset a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons? Dataset: https://github.com/MasteriNeuron/datasets.git (Include your Python code and output in the code box below.)


import pandas as pd
import matplotlib.pyplot as plt

# Load dataset from local path or download beforehand
df = pd.read_csv('googleplaystore.csv')

# Data Cleaning: Remove rows with missing ratings
df_clean = df.dropna(subset=['Rating'])

# Convert ratings to numeric (if needed)
df_clean['Rating'] = pd.to_numeric(df_clean['Rating'], errors='coerce')

# Calculate average rating per category
category_ratings = df_clean.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# Display top 5 and bottom 5 categories by avg rating
print("Top 5 categories by average rating:")
print(category_ratings.head())

print("\nBottom 5 categories by average rating:")
print(category_ratings.tail())

# Visualization
plt.figure(figsize=(12,6))
category_ratings.plot(kind='bar', color='skyblue')
plt.title('Average App Rating by Category (Google Play Store)')
plt.ylabel('Average Rating')
plt.xlabel('App Category')
plt.ylim(1,5)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
